### Task 1

```python
four_regions = pd.read_csv("Data/Lab2/Data Geographies - v1 - by Gapminder - list-of-countries-etc.csv")
income = pd.read_csv("Data/Lab2/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
life_expect = pd.read_csv("Data/Lab2/life_expectancy_years.csv")
pop = pd.read_csv("Data/Lab2/population_total.csv")
```

```python
income_2010 = income[["country", "2010"]]
income_2010 = income_2010.rename(columns = {"country":"country", "2010": "in_10"})

life_expect_2010 = life_expect[["country", "2010"]]
life_expect_2010 = life_expect_2010.rename(columns = {"country":"country", "2010": "life_ext_10"})

pop_2010 = pop[["country", "2010"]]
pop_2010 = pop_2010.rename(columns = {"country":"country", "2010": "pop_10"})

four_regions = four_regions[["name", "four_regions"]]
four_regions = four_regions.rename(columns = {"name":"country", "four_regions": "four_regions"})
four_regions
```

```python
pop_2010.pop_10 = pop_2010['pop_10'].replace({'k': '*1e3', 'M': '*1e6', 'B': '1e9'}, regex=True).map(pd.eval).astype(int)
income_2010.in_10 = income_2010['in_10'].replace({'k': '*1e3', 'M': '*1e6', 'B': '1e9'}, regex=True).map(pd.eval).astype(int)
```

```python
ex1data = four_regions.merge(income_2010, on = 'country')
ex1data = ex1data.merge(life_expect_2010, on = 'country')
ex1data = ex1data.merge(pop_2010, on = 'country')
ex1data
```

```python
plt.figure(figsize=(12, 6))
plot1 = sns.scatterplot(data=ex1data, x="in_10", y="life_ext_10", size="pop_10", hue = "four_regions", 
                        palette = pal_color, alpha=0.7, linewidth = 1, sizes=(50, 1500))
plot1.set(xlabel = "Income", ylabel="Life Expectancy", xscale= "log", ylim=(10,90))
plot1.set_xticks([400, 500, 1000, 2000, 4000, 8000, 16000, 32000, 64000])
plot1.set_xticklabels(["0", "500", "1000", "2000", "4000", "8000", "16k", "32k", "64k"])
plt.grid(alpha=0.2) 
plt.text(0.15, 0.28, '2010', fontsize = 180, alpha=0.1, transform=plot1.transAxes)

plt.legend(bbox_to_anchor=(1, 1))
```

### Task 2

```python
exports = pd.read_csv("Data/Lab2/exports_percent_of_gdp.csv")
imports = pd.read_csv("Data/Lab2/imports_percent_of_gdp.csv")
energy = pd.read_csv("Data/Lab2/energy_use_per_person.csv")
```

```python
exports_1997 = exports[["country", "1997"]]
exports_1997 = exports_1997.rename(columns = {"country":"country", "1997": "export_97"})

imports_1997 = imports[["country", "1997"]]
imports_1997 = imports_1997.rename(columns = {"country":"country", "1997": "import_97"})

energy_1997 = energy[["country", "1997"]]
energy_1997 = energy_1997.rename(columns = {"country":"country", "1997": "energy_97"})
```

```python
energy_1997.fillna('0', inplace=True)
energy_1997
```

```python
energy_1997.energy_97 = energy_1997['energy_97'].replace({'k': '*1e3', 'M': '*1e6', 'B': '1e9'}, regex=True).map(pd.eval).astype(int)
```

```python
ex2data = four_regions.merge(exports_1997, on = 'country')
ex2data = ex2data.merge(imports_1997, on = 'country')
ex2data = ex2data.merge(energy_1997, on = 'country')
ex2data
```

```python
plt.figure(figsize=(12, 6))
plot2 = sns.scatterplot(data=ex2data, x="export_97", y="import_97", size="energy_97", hue = "four_regions", 
                        palette = pal_color, alpha=0.7, linewidth = 1, sizes=(20, 800), legend = False)
plot2.set(xlabel = "Exports (\% of GDP)", ylabel="Imports (\% of GDP)", xlim = (0,220), ylim=(-20,400))
plot2.set_xticks([20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240])
plt.text(0.08, 0.25, '1997', fontsize = 180, alpha=0.1, transform=plot2.transAxes)

plt.grid(alpha=0.2) 
```
